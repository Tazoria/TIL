### 1Ô∏è‚É£ ÌñâÏúºÎ°ú Ïó∞Í≤∞ÌïòÍ∏∞

```python
# pd.concat
df1 = pd.DataFrame([['a', 1], ['b', 2]], columns=['letter', 'number'])
df2 = pd.DataFrame([['c', 3], ['d', 4]], columns=['letter', 'number'])
df3 = pd.DataFrame([['w', 5, '!'], ['f', 6, '@']], columns=['letter', 'number', 'etc'])

print('[df1]\n', df1, '\n')
print('[df2]\n', df2, '\n')
print('[df3]\n', df3, '\n')
```

```python
# Ïª¨ÎüºÎ™Ö Í∏∞Ï§ÄÏúºÎ°ú Ïó∞Í≤∞
df_rowconcat = pd.concat([df1, df2, df3])
df_rowconcat
```

```python
# Í≥µÌÜµÎêú Ïª¨ÎüºÎßå ÎÇ®Í∏∞Í∏∞
df_rowconcat2 = pd.concat([df1,df2,df3], join='inner')
df_rowconcat2
```

```python
# Ïù∏Îç±Ïä§ Ïû¨ÏßÄÏ†ï
df_rowconcat3 = pd.concat([df1, df2, df3], join='inner', ignore_index=True)
df_rowconcat3
```

### 2Ô∏è‚É£ Ïó¥Î°ú Ïó∞Í≤∞ÌïòÍ∏∞

```python
df4 = pd.DataFrame({'age':[20,21,22]}, index=['amy','james','david'])
df5 = pd.DataFrame({'phone':['010-1111-1111', '010-2222-2222', '010-3333-3333']}, index=['amy','james','david'])
df6 = pd.DataFrame({'job':['studnet','programer','ceo','designer']}, index=['amy','james','david','kim'])

print('[df4]\n', df4, '\n')
print('[df5]\n', df5, '\n')
print('[df6]\n', df6, '\n') # kim Ìïú ÎßàÎ¶¨ Ï∂îÍ∞Ä
```

```python
# Î≥ÑÎã§Î•∏ ÏòµÏÖò ÏóÜÏù¥ Î∂ôÏù¥Í∏∞ : Î™®Îì† Ïª¨ÎüºÏùÑ Îã§ Ìè¨Ìï®(Ìï¥Îãπ Ïª¨ÎüºÏù¥ ÏóÜÎäî Í≤ΩÏö∞ NaNÍ∞í), axis=0Î∞©Ìñ•ÏúºÎ°ú Ïù¥Ïñ¥Î∂ôÏûÑ
df_column_concat = pd.concat([df4,df5,df6])
df_column_concat
```

```python
# joiun='inner': Ìï¥Îãπ Îç∞Ïù¥ÌÑ∞Ïóê Í≥µÌÜµÎêú Î∂ÄÎ∂ÑÎßå Î≥ëÌï©
# axis=1: Ïª¨ÎüºÎ∞©Ìñ• Î≥ëÌï©
df_column_concat2 = pd.concat([df4,df5,df6], join='inner', axis=1)
df_column_concat2
```

## üí° merge
- pd.merge( left, right, on = Í∏∞Ï§ÄÏª¨Îüº, how = Ïó∞Í≤∞Î∞©Î≤ï )

```python
df = pd.read_csv('/content/drive/MyDrive/class/·Ñë·Ö°·Ü´·ÑÉ·Ö°·Ñâ·Ö≥·Ñâ·Öµ·ÜØ·Ñâ·Ö≥·Ü∏/scores.csv')
df.head(3)
```

```python
df7 = df.loc[[1,2,3]][['name', 'eng']]
df8 = df.loc[[1,2,4]][['name', 'math']]

print('[df7]\n', df7, '\n')
print('[df8]\n', df8, '\n')
```

```python
# Í≥µÌÜµ Îç∞Ïù¥ÌÑ∞ÎßåÏúºÎ°ú Ïó∞Í≤∞
pd.merge(df7, df8, on='name') # default: how='inner'
```

```python

pd.merge(df7, df8, on='name', how='outer') 
```

```python
# how='left': ÏôºÏ™Ω ÌîÑÎ†àÏûÑÏùÑ Í∏∞Ï§ÄÏúºÎ°ú Ïò§Î•∏Ï™Ω ÌîÑÎ†àÏûÑÏùÑ Î≥ëÌï©(Ïò§Î•∏Ï™ΩÏóêÎßå ÏûàÎäî Í≤ΩÏö∞Îäî Ïß§Î¶º)
pd.merge(df7, df8, on='name', how='left') 
```

```python
# how='right': Ïò§Î•∏Ï™Ω ÌîÑÎ†àÏûÑÏùÑ Í∏∞Ï§ÄÏúºÎ°ú ÏôºÏ™Ω ÌîÑÎ†àÏûÑÏùÑ Î≥ëÌï©(ÏôºÏ™ΩÏóêÎßå ÏûàÎäî Í≤ΩÏö∞Îäî Ïß§Î¶º)
pd.merge(df7, df8, on='name', how='right') 
```

## üí° Ïó¥ÏùÑ ÌñâÏúºÎ°ú Î≥¥ÎÇ¥Í∏∞

### 1Ô∏è‚É£ pivot
- Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ.pivot(index = Ïù∏Îç±Ïä§Î°ú ÏÇ¨Ïö©Ìï† Ïª¨Îüº, columns=Ïª¨ÎüºÏúºÎ°ú ÏÇ¨Ïö©Ìï† Ïª¨Îüº, values=Í∞íÏúºÎ°ú ÏÇ¨Ïö©Ìï† Ïª¨Îüº)


```python
df = df.melt(id_vars='name', var_name='subject',
             value_name='score')
df.head(10)
```

```python
def get_grade(x):
  if x >= 90:
    grade = 'A'
  elif x >= 80:
    grade = 'B'
  elif x >= 70:
    grade = 'C'
  elif x >= 60:
    grade = 'D'
  else:
    grade = 'F'
  return grade
```

```python
df['grade'] = df['score'].apply(get_grade)
df = df.sort_values('name')
df.head(10)
```

```python
# pivot
df.pivot(index='name', columns='subject', values='score').head(10)
```

```python
df.pivot(index='name', columns='subject', values='grade').head(10)
```

```python
df.pivot(index='name', columns='subject', values=['score', 'grade']).head(10)
```
