```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age', 'fare']]
df.head()
```

## 1Ô∏è‚É£ applymap ( )

```python
# Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏóê Ïó∞ÏÇ∞Ï†ÅÏö©(Í∞Å ÏõêÏÜåÏóê + 10ÏùÑÌïòÎäî Ìï®Ïàò)
def add_10(n):
  return n+10

df_map = df.applymap(add_10)
print(df_map.head(3))
```

## 2Ô∏è‚É£ apply ( )

```python
# Í≤∞Ï∏°Ïπò Ï∞æÏïÑÏ£ºÎäî Ìï®Ïàò
def missing_value(series):
  return series.isnull()
```

```python
result = df.apply(missing_value, axis=0)
print(result.head(3))
```

```python
def min_max(x):
  return x.max() - x.min()
result2 = df.apply(min_max)
print(result2)
```

## ‚úÖ Ïã§Ïäµ2 : titanic
- age :  ÌèâÍ∑†ÏúºÎ°ú ÎåÄÏ≤¥
- embarked : ÏµúÎπàÍ∞íÏúºÎ°ú ÎåÄÏ≤¥
- embark_town : Í≤∞Ï∏°Í∞íÏù¥ ÏûàÎäî Ìñâ ÏÇ≠Ï†ú


```python
# ageÏùò Í≤∞Ï∏°ÏπòÎ•º ÌèâÍ∑†ÏúºÎ°ú ÎåÄÏ≤¥ÌïòÍ∏∞
df = titanic.copy()
df.head()
```

```python
mean_age = df['age'].mean().round(2)
df['age'].fillna(mean_age, inplace=True)
df.info()
```

```python
# embarkedÏùò Í≤∞Ï∏°ÏπòÎ•º ÏµúÎ∞òÍ∞íÏúºÎ°ú ÎåÄÏ≤¥
mode_embarked = df['embarked'].mode()[0]
df['embarked'].fillna(mode_embarked, inplace=True)
df.info()
```

```python
# embark_town Í≤∞Ï∏°Í∞íÏù¥ ÏûàÎäî Ìñâ ÏÇ≠Ï†ú
df = df.dropna(subset=['embark_town'], how='any', axis=0)
df.info()
```

```python
# deck Ïª¨Îüº ÏÇ≠Ï†ú
df.drop('deck', inplace=True, axis=1)
df.info()
```

```python
# titanic['adult/child'] Ïª¨ÎüºÏñº ÎßåÎì§Í≥† 20ÏÇ¥ Ïù¥ÏÉÅÏù¥Î©¥ adult Ïù¥ÌïòÎ©¥ childÎ°ú Íµ¨Î∂Ñ
df['adult/child'] = titanic.age.apply(lambda x: 'adult' if x> 20 else 'child')
df.head()
```
