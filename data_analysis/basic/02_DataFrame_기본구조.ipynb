```python
# 실습 1
data = {'국어':[80, 90, 70, 30], '영어':[90, 70, 60, 40], '수학':[90, 60, 80, 70]}

# 행에 학생 이름 넣기 : 춘향, 몽룡, 향단, 방자
students = ['춘향', '몽룡', '향단', '방자']
df = pd.DataFrame(data, index=students)
print(df)

# indexing으로 몽룡이의 국어점수 뽑아보기
print('몽룡이의 국어점수: ', df['국어'][1])
print('몽룡이의 국어점수: ', df['국어']['몽룡'])
```

### 💡 iloc, loc을 통한 인덱싱

```python
# 실습 2

# numpy를 사용해서 random.randint()사용해서 60~100사이 12개 숫자만들고 reshape()
# index = ['James', 'Maria', 'Brian', 'Elice']
# columns = ['kor', 'eng', 'math']

np.random.seed(2023)
score = np.random.randint(60, 100+1, 12).reshape(4, 3)
index = ['James', 'Maria', 'Brian', 'Elice']
columns = ['kor', 'eng', 'math']
score_df = pd.DataFrame(score, index=index, columns=columns)
score_df
```

```python
# select row
# row index를 사용하여 행 1개를 선택
df3 = score_df[:]
mar1 = df3.loc['Maria'] # location
mar2 = df3.iloc[1] # integer location 

print(mar1)
print(mar2)
print(mar1 == mar2)
```

```python
label = df3.loc[['Brian', 'Elice']]
label2 = df3.iloc[[2,3]]
print(label)
print(label2)
```

```python
# 실습 3
exam_data = {'수학':[90,80,70],
             '영어':[98,89,95],
             '음악':[85,95,100],
             '체육':[100,90,90]}
exam_data = pd.DataFrame(exam_data, index=['길동','철수','영희'])
exam_data
df = exam_data.copy()
df
```

### 💡 reindex( )

```python
# reindex
dict_data = {'c0':[1,2,3], 'c1':[4,5,6], 'c2':[7,8,9],
             'c3':[10,11,12], 'c4':[13,14,15]}
dict_data = pd.DataFrame(dict_data, index=['r0', 'r1', 'r2'])
dict_data
```

```python
# 인덱스를 새로 지정해서 행 추가
new_index = ['r0', 'r1', 'r2','r3','r4']
ndf = dict_data.reindex(new_index)
print(ndf)
```

```python
# reindex로 발생한 NaN값을 0으로 채우기
ndf2 = dict_data.reindex(new_index, fill_value=0)
```

```python
# reset_index
ndf2 = ndf2.reset_index()
ndf2
```

## 8️⃣ 데이터 연산

### 🔘 시리즈

```python
# 시리즈 선언
student1 = pd.Series({'국어':100, '영어':80, '수학':90})
student2 = pd.Series({'수학':80, '국어':90, '영어':80})

# 과목별로 점수를 200으로 나누기
percentage = student1 / 200
print(percentage, type(percentage))



# 시리즈 사칙연산 1 : 시리즈간
# 결측치가 없는 경우
add = student1 + student2
sub = student1 - student2
mul = student1 * student2
div = student1 / student2

result = pd.DataFrame([add, sub, mul, div], index=['덧셈', '뺄셈', '곱셈', '나눗셈'])
print(result)

# 결측치가 있는경우 - 결측치 + 값 = 결측치
student1 = pd.Series({'국어':np.nan, '영어':80, '수학':90})
student2 = pd.Series({'수학':80, '국어':90, '영어':80})

add = student1 + student2
sub = student1 - student2
mul = student1 * student2
div = student1 / student2

result = pd.DataFrame([add, sub, mul, div], index=['덧셈', '뺄셈', '곱셈', '나눗셈'])
print(result)
```

```python
# 시리즈 사칙연산2 : 연산메소드 사용
add = student1.add(student2, fill_value=0) # 덧셈
sub = student1.sub(student2, fill_value=0) # 뺄셈
mul = student1.mul(student2, fill_value=0) # 곱셈
div = student1.div(student2, fill_value=0) # 나눗셈

result = pd.DataFrame([add, sub, mul, div], index=['덧셈', '뺄셈', '곱셈', '나눗셈'])
result
```
