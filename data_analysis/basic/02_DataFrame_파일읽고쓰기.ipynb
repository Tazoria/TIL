## ‚úÖ Ïã§Ïäµ : auto-mpg.csv

```python
# Ïã§Ï†ú Îç∞Ïù¥ÌÑ∞ ÌÉêÏÉâ(exploratory_analysis)
# header=NoneÏòµÏÖòÏùÑ Ï§òÏïºÏßÄ Îß® ÏúÑÏùò Îç∞Ïù¥ÌÑ∞ ÏÜêÏã§Ïù¥ ÏïàÎê®.
car = pd.read_csv('/content/drive/MyDrive/class/·Ñë·Ö°·Ü´·ÑÉ·Ö°·Ñâ·Ö≥data/auto-mpg.csv', header=None)
car.head()
```

```python
'''
< ÌîºÏ≥ê >
1) mpg: Ïó∞ÎπÑ
2) cylinders: Ïã§Î¶∞Îçî Í∞úÏàò
3) displacement: Î∞∞Í∏∞Îüâ
4) horsepower: ÎßàÎ†•
5) weight: Î¨¥Í≤å
6) acceleration: ÏóîÏßÑÏù¥ Ï¥àÎãπ ÏñªÏùÑ Ïàò ÏûàÎäî Í∞ÄÏÜçÎ†•
7) model year: Ï∂úÏãú ÎÖÑÎèÑ
8) origin: Ï†úÏ°∞ Ïû•ÏÜå(1: ÎØ∏Íµ≠ USA, 2: Ïú†ÎüΩ EU, 3: ÏùºÎ≥∏ JPN)
9) name: ÏûêÎèôÏ∞® Ïù¥Î¶Ñ
'''
# Ïª¨ÎüºÏù¥ Îπ†Ï†∏ÏûàÎã§! - Ïª¨Îüº ÏßÄÏ†ïÌï¥Ï£ºÍ∏∞
car.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight',
              'acceleration', 'model year', 'origin', 'name']
df = car.copy()
df.head()
```
